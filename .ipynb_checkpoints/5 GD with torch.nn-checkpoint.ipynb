# Gradient descent using Pytorch neural network features

```python
# Import packages
import numpy as np
import torch
from gdtnn import *
```

```python
# Make data
n_data = 101
x = torch.linspace(-1, 1, n_data).reshape(n_data, 1)

# Target functions, from easy to hard
# y = torch.maximum(2 * x, torch.zeros_like(x))
# y = x**3
y = np.sin(np.pi * x)

# Get cpu or gpu device for training
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using {} device".format(device))
x, y = x.to(device), y.to(device)

# Make neural net with set number of neurons in hidden layer
model = ToyNN(n_hidden=10).to(device)
print(model)

grad_desc(x, y, model, n_data, lr=1e-1, max_it=int(2e3), plot_every=100)
```
